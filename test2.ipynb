```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('test.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Normalize the image
image = image/255.0
```

```python
reduce_factor = 0.5
```

```python
image = cv2.resize(image, (0,0), fx=reduce_factor, fy=reduce_factor)
```

```python
plt.imshow(image)
```

```python
def conv(image, kernel):
    new_image = np.zeros((image.shape[0] - 2, image.shape[1] - 2))
    for i in range(image.shape[0] - 2):
        for j in range(image.shape[1] - 2):
            # for k in range(3):
            #     new_image[i, j, k] = np.sum(image[i:i+3, j:j+3, k] * kernel)
            new_image[i, j] = np.sum(image[i:i+3, j:j+3] * kernel)
    return new_image
```

```python
edge_detection_kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
sharp_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
blur_kernel = np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]])/9.0
left_sobel_kernel = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
right_sobel_kernel = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
top_sobel_kernel = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
bottom_sobel_kernel = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

kernels = [edge_detection_kernel, sharp_kernel, blur_kernel, left_sobel_kernel, right_sobel_kernel, top_sobel_kernel, bottom_sobel_kernel]
kernel_names = ['edge_detection_kernel', 'sharp_kernel', 'blur_kernel', 'left_sobel_kernel', 'right_sobel_kernel', 'top_sobel_kernel', 'bottom_sobel_kernel']
```

```python
for i in range(len(kernels)):
    new_image = conv(image, kernels[i])
    plt.figure()
    plt.imshow(new_image)
    plt.title(kernel_names[i])
```
