```python
import cv2
import rasterio
from rasterio.plot import show
import numpy as np
import matplotlib.pyplot as plt
```

```python
park = rasterio.open('park.tif')
```

```python
print(park.meta)
```

```python
red = park.read(1)
green = park.read(2)
blue = park.read(3)

rgb_img = np.dstack((red, green, blue))
```

```python
plt.imshow(rgb_img)
```

```python
# imwrite 
bgr_image = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2BGR)
cv2.imwrite('park.png', bgr_image)
```

```python
#show(rgb_img, cmap='terrain')
show(rgb_img)
```
