```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
img.shape
```

```python
filter = np.array([[255, 255, 255], 
                    [0, 0, 0], 
                    [255, 255, 255]])

plt.imshow(filter, cmap='gray')
```

```python
img.shape
```

```python
filter.shape
```

```python
def conv(img, filter):
    new_image_height = img.shape[0] - filter.shape[0] + 1
    new_image_width = img.shape[1] - filter.shape[1] + 1
    new_image = np.zeros((new_image_height, new_image_width))
    for r in range(new_image_height):
        for c in range(new_image_width):
            curr_region = img[r:r+filter.shape[0], c:c+filter.shape[1]]
            curr_result = curr_region * filter
            conv_sum = np.sum(curr_result)
            new_image[r, c] = conv_sum
    return new_image

```

```python
new_image = conv(img, filter)
new_image2 = conv(new_image, filter)
new_image3 = conv(new_image2, filter)
new_image4 = conv(new_image3, filter)
new_image5 = conv(new_image4, filter)
plt.imshow(new_image, cmap='gray')
```
